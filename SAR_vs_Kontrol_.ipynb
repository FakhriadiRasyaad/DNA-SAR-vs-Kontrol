# Data Asli

```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/DNA/Code + Dataset/Data Metilasi DNA SAR.xlsx"

df = pd.read_excel(path)
df = df.applymap(lambda x: x.strip() if isinstance(x,str) else x)

df = df.replace(["na","NA","Na",""], np.nan)

for col in ["Posisi 1","Posisi 2","Posisi 3","Posisi 4","Posisi 5","Posisi 6"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[:]
```

```python
use_cols = ["Posisi 1", "Posisi 2", "Posisi 3", "Posisi 4", "Posisi 5", "Posisi 6", "Kelas"]
df2 = df[use_cols]
```

```python
df2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in ["Posisi 1", "Posisi 2", "Posisi 3", "Posisi 4", "Posisi 5", "Posisi 6"]:
    ax = sns.boxplot(x="Kelas", y=col, data=df2)

    # Ubah label sumbu
    ax.set_xlabel("Class")
    ax.set_ylabel("Methylation (%)")
    ax.set_title(col.replace("Posisi", "Position"))

    # Ubah label tick kelas
    ax.set_xticklabels(["SAR", "Control"])

    plt.show()
```

```python
use_cols = ["Posisi 1","Posisi 2","Posisi 3","Posisi 4", "Posisi 5", "Posisi 6", "Kelas"]
df2 = df[use_cols]

X = df2.drop(columns="Kelas")
y = df2["Kelas"]
```

```python
X_imp = X.copy()
for col in X_imp.columns:
    X_imp[col] = X_imp[col].fillna(X_imp[col].mean())

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imp)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

param_grid = [
    # Logistic Regression
    {
        "model": [LogisticRegression(solver="liblinear")],
        "model__C": [0.1, 1, 10]
    },
    # SVM
    {
        "model": [SVC(probability=True)],
        "model__C": [0.1, 1, 10],
        "model__kernel": ["linear", "rbf"]
    },
    # KNN
    {
        "model": [KNeighborsClassifier()],
        "model__n_neighbors": [3, 5, 7]
    }
]
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([("model", LogisticRegression())])

grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring="roc_auc",
    cv=cv,
    n_jobs=-1
)

grid.fit(X_scaled, y)

print("Best params:", grid.best_params_)
print("Best ROC-AUC:", grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix

best_model = grid.best_estimator_

y_pred = cross_val_predict(best_model, X_scaled, y, cv=cv)

print(classification_report(y, y_pred))
print("Confusion matrix:\n", confusion_matrix(y, y_pred))
```

# Augmentasi Data

```python
import numpy as np
import pandas as pd

X_num = X_imp.values
y_num = y.values

X_ctrl = X_num[y_num=="Kontrol"]
X_sar  = X_num[y_num=="SAR"]
```

```python
mu_ctrl  = X_ctrl.mean(axis=0)
cov_ctrl = np.cov(X_ctrl, rowvar=False)

mu_sar  = X_sar.mean(axis=0)
cov_sar = np.cov(X_sar, rowvar=False)
```

```python
np.random.seed(42)

n_new = 100

X_ctrl_syn = np.random.multivariate_normal(mu_ctrl, cov_ctrl, n_new)
X_sar_syn  = np.random.multivariate_normal(mu_sar,  cov_sar,  n_new)

X_ctrl_syn = np.clip(X_ctrl_syn, 0, 100)
X_sar_syn  = np.clip(X_sar_syn,  0, 100)
```

```python
X_all = np.vstack([X_ctrl, X_sar, X_ctrl_syn, X_sar_syn])
y_all = (["Kontrol"]*len(X_ctrl) +
         ["SAR"]*len(X_sar) +
         ["Kontrol"]*n_new +
         ["SAR"]*n_new)
```

```python
df3 = pd.DataFrame(X_all, columns=["Posisi 1","Posisi 2","Posisi 3","Posisi 4", "Posisi 5", "Posisi 6"])
df3["Kelas"] = y_all
```
